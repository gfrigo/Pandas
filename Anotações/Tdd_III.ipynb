# Pandas 

#### Imports e Dataframe

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
```

```python
df.columns
```

```python
df1 = df[['ESCMAE', 'IDADEMAE', 'IDADEPAI', 'QTDFILVIVO', 'QTDFILMORT']]
df1.head()
```

```python
def top(df, n=5, column='QTDFILVIVO'):
    return df.fillna(0).sort_values(by=column, ascending=False)[:n]
```

```python
def amplitude(array):
    return array.max() - array.min()
```

#### Groupby

Neste tipo de tabela as funções são orgnaizadas como ".função()" podendo ter inúmeras.

```python
df.groupby(['DTNASC', 'SEXO'])['IDADEMAE'].count().to_frame()
```

```python
df.groupby(['DTNASC', 'SEXO'])['IDADEMAE'].count().unstack()
```

#### Mapping

Mapear valores e renomear categorias. Para este método seguir os passos abaixo:

```python
df_mapping = df.set_index('ESCMAE')
df_mapping.head()
```

```python
mapping = {
    'Nenhuma' : 'Nenhuma',
    '1 a 3 anos' : '1 a 7 anos',
    '4 a 7 anos':'1 a 7 anos',
    '8 a 11 anos': '8 anos ou mais',
    '12 anos ou mais': '8 anos ou mais'
}
```

```python
df_mapping.groupby(mapping).sum()
```

#### aggregate() ou agg()

Este método aplica funções em uma tabela groupby.

```python
df_agg = df[['ESCMAE', 'IDADEMAE', 'IDADEPAI']]
```

Formas diferentes de chamada de funções:

```python
df_agg.groupby('ESCMAE').aggregate(['min', np.median, max])
# df_agg.groupby('ESCMAE').agg(['max', np.median, max])
```

Renomeando funções na tabela

```python
df_agg.groupby('ESCMAE').aggregate([('minimo', 'min'), ('mediana', np.median), ('maximo', max)])
```

Passar funções para colunas específicas:

```python
df_agg.groupby('ESCMAE').aggregate({
                                    'IDADEMAE': [('minimo', 'min'), ('maximo', max)],
                                    'IDADEPAI': [('contagem', 'count')]
                                   })
```

Aplicar funções difinadas pelo programador:

```python
df_agg.groupby(['ESCMAE']).agg(amplitude)
```

#### apply()

O método apply() tem como finalidade a aplicação de funções definidas pelo programador em tabelas groupby.

```python
df_apply = df1
```

```python
df_apply.groupby('ESCMAE').apply(top)
```

```python
df_apply.groupby('ESCMAE').apply(top, n=6, column='QTDFILMORT')
```

### Tabela Pivô e Cut

Forma mais organizada e estruturada do Groupby

```python
pd.pivot_table(data=df,
               values='IDADEMAE',
               index='ESCMAE',
               aggfunc='sum',
               margins=True,
               margins_name='Total'
               )
```

O cut servirá para criar categorias personalizadas dos nossos valores onde nós mesmos passamos os valores de quebra da classe:

```python
idade_mae = pd.cut(df['IDADEMAE'], [0,18,80])
idade_mae.value_counts()
```

Já o qcut passaremos a quantidade de quartis para quebrar em categorias:

```python
idade_pai = pd.qcut(df['IDADEPAI'], 2)
idade_pai.value_counts()
```

Tabela pivô estruturada

```python
df.pivot_table(index=['ESCMAE', idade_mae],
               values=['PESO','IDADEPAI'],
               columns=[idade_pai],
               aggfunc={
                   'PESO': [('média', 'mean'), ('mediana', 'median')],
                   'IDADEPAI': [('máximo', np.max), ('mínimo', np.min)]
               },
               dropna=True)
```
